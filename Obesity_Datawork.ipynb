# Obesity Data Analysis

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm 
from scipy import stats
from math import comb, factorial
sns.set()
```

```python
Data = pd.read_csv('Obesity_Dataset.csv')
Data
```

## Inspect Data

```python
Data.info()
```

```python
sns.histplot(Data['Age'], bins=10, kde=True, color='red')
plt.title('Distribution Of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show
```

```python
sns.boxplot(x='Height', y='Age', data=Data, palette='Set3')
plt.show()
```

```python
sns.scatterplot(Data, x='Age', y='Physical_Excercise', palette='coolwarm', hue='Age')
plt.figure(figsize=(10,5))
plt.title('Age vs Physical_Excercise')
plt.xlabel('Age')
plt.ylabel('Physical_Excercise')
plt.show()
```

```python
sns.scatterplot(Data, x='Age', y='Number_of_Main_Meals_Daily', palette='coolwarm', hue='Age')
plt.figure(figsize=(10,5))
plt.title('Age vs Number_of_Main_Meals_Daily')
plt.xlabel('Age')
plt.ylabel('Number_of_Main_Meals_Daily')
plt.show()
```

```python
sns.scatterplot(Data, x='Age', y='Height', palette='coolwarm', hue='Age')
plt.figure(figsize=(10,5))
plt.title('Age vs Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

```python
sns.scatterplot(Data, x='Age', y='Frequency_of_Consuming_Vegetables', palette='coolwarm', hue='Age')
plt.figure(figsize=(10,5))
plt.title('Age vs Frequency_of_Consuming_Vegetables')
plt.xlabel('Age')
plt.ylabel('Frequency_of_Consuming_Vegetables')
plt.show()
```

```python
correlation, p_value = stats.pearsonr(Data['Age'], Data['Height'])
correlation, p_value
```

```python
correlation_matrix = Data.corr()
correlation_matrix
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(Data.corr(), annot=True, cmap='coolwarm', annot_kws={'size':10},
xticklabels={'size':14},
yticklabels={'size':14})
plt.title('Correlation heatmap')
plt.show()
```

```python
X = Data['Age']
Y = Data['Height']
```

```python
X = sm.add_constant(X)

model = sm.OLS(Y,X).fit()
```

```python
model.summary()
```

```python
X = Data['Age']
Y = Data['Height']

beta_0 = 164.6385 #(Inercept)
beta_1 = 0.0937 #(Slope)

y_pred = beta_0 + beta_1*X

sns.scatterplot(x = X, y=Y)
plt.title('Age Vs. Height')

plt.plot(X, y_pred, color = 'red', label = 'Trendline')

plt.legend()

plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```
